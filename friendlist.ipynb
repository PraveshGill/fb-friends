```python
browser.get('https://m.facebook.com/your_fb_username/friends')
```

```python
more_friends = browser.find_element_by_id('m_more_friends')

more_friends.click()
#browser.set_window_position() to scroll down the paga
```

```python
html=browser.page_source
finallist=[]
```

```python
html
```

```python
soup=BeautifulSoup(html,"html5lib")
```

```python
print(soup.prettify())
```

```python
table=soup.findAll('table')
```

```python
print(table[2].prettify())
#from table no 2 to 25
```

```python
for i in range(2,26):
    l=[]
    a=table[i].tbody.tr.a.text
    b=table[i].tbody.tr.div.text
    l.append(a)
    l.append(b)
    finallist.append(l)
```

```python
finallist
```

```python
from tabulate import tabulate
```

```python
print(tabulate(finallist,tablefmt="grid",headers=["Name","No. of mutual friend"]))
```
